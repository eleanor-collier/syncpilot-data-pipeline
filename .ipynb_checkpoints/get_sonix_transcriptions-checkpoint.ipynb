## Upload disclosure recordings to Sonix

```python
# Calculate total number of pages on Sonix to loop through
url = "https://api.sonix.ai/v1/media"
response = requests.get(url, headers=headers).json()
n_pages = response['total_pages']
```

```python
# Get names of recordings already uploaded to Sonix
recording_names = []
for page in range(n_pages + 1):
    url = "https://api.sonix.ai/v1/media?page=" + str(page)
    response = requests.get(url, headers=headers).json()
    recording_names_on_page = [recording.get('name') for recording in response['media']]
    recording_names.extend(recording_names_on_page)
```

```python
url = "https://api.sonix.ai/v1/media"

# Loop through participant folders & self-disclosure recordings on Google Drive
for subdir, dirs, files in os.walk('Data/recordings'):
    for file in files: 
        
        # Only include real participants' data and self (not partner's) disclosures
        if '/P' in os.path.join(subdir, file) and 'self' in os.path.join(subdir, file):
            # Get recording name from participant ID and disclosure name
            file_path = os.path.join(subdir, file)
            regex = re.compile('/')
            recording_name = regex.split(file_path)[2] + "." + regex.split(file_path)[3].replace(".m4a","")
            
            # Exclude recordings previously uploaded to Sonix
            if recording_name not in recording_names:
                
                # Upload recording
                files = {
                    "file": open(file_path, "rb")
                } 
                params = {
                    "language": "en",
                    "name": recording_name
                }
                print("Uploading " + recording_name + " to Sonix")
                response = requests.post(url, headers=headers, files=files, params=params)
                
                # Stop if upload is not successful
                if str(response) != '<Response [200]>':
                    print("Error: " + response.text)
                    break
                
```

## Download transcriptions

```python
# Calculate total number of pages on Sonix to loop through
url = "https://api.sonix.ai/v1/media"
response = requests.get(url, headers=headers).json()
n_pages = response['total_pages']
```

```python
# Get list of all ids for recordings successfully transcribed on Sonix
sonix_ids = []
for page in range(n_pages + 1):
    url = "https://api.sonix.ai/v1/media?page=" + str(page)
    response = requests.get(url, headers=headers).json()
    sonix_ids_on_page = [recording.get('id') for recording in response['media'] 
                         if recording.get('status')=='completed']
    sonix_ids.extend(sonix_ids_on_page)
```

```python
# Loop through all recordings and get transcripts
all_transcripts = []
for sonix_id in sonix_ids:
    
    # Get json file from Sonix
    url = "https://api.sonix.ai/v1/media/" + sonix_id + "/transcript.json"
    response = requests.get(url, headers=headers).json()
    
    # Get participant ID
    regex = re.compile('\.')
    ID = regex.split(response['name'])[0].replace("P","")
    
    # Get disclosure name
    regex = re.compile('_+')
    disclosure = regex.split(response['name'])[-1]
    
    # Get transcript + time stamps
    transcript = pd.concat(
        [pd.DataFrame(speaker.get('words')) for speaker in response['transcript']], 
        ignore_index=True
    )
    transcript_df = transcript.drop(['highlight', 'strikethrough'], axis=1)
    
    # Add participant ID + disclosure name to transcript
    transcript_df.insert(0, 'ID', ID)
    transcript_df.insert(1, 'disclosure', disclosure)
    
    # Append to master data frame
    all_transcripts.append(transcript_df)

# Concatenate list into dataframe
all_transcripts_df = pd.concat(all_transcripts, ignore_index=True)
```

```python
# Save as csv
all_transcripts_df.to_csv('Data/processed/transcripts_by_word.csv', index=False)
```

### Group words by moving window

#### 15 second window

```python
window_size = 15
all_transcripts_by_window = []
for transcript in all_transcripts:
    last_word_time = math.ceil(transcript.end_time.iat[-1])
    transcript_by_window = []
    for time in range(window_size, last_word_time+1):
        rows = transcript.loc[(transcript['end_time'] > time - window_size) & (transcript['end_time'] <= time)]
        words = rows.text.str.cat(sep=' ')
        transcript_window = pd.DataFrame({
            'ID': [transcript.ID[1]],
            'disclosure': [transcript.disclosure[1]],
            'time': time,
            'text': words
        })
        transcript_by_window.append(transcript_window)
    transcript_by_window_df = pd.concat(transcript_by_window, ignore_index=True)
    all_transcripts_by_window.append(transcript_by_window_df)

# Concatenate list into dataframe
all_transcripts_by_window_df = pd.concat(all_transcripts_by_window, ignore_index=True)
```
